# Zadanie 9 – Pipeline z wieloma krokami

Stwórz 4 komponenty:
- FiltrowanieDodatnich  – usuwa liczby ujemne
- Zaokraglanie  – zaokrągla do całkowitych
- Skalowanie  – mnoży przez stały współczynnik
- Sumowanie  – zwraca sumę listy

Stwórz klasę  Pipeline  i przetestuj na danych:  [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

```python
# Data
data = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

class FiltrowanieDodatnich:
    def __init__(self):
        self.name = "Filtrowanie liczb dodatnich..."

    def wykonaj(self, data):
        print(self.name)
        return [x for x in data if x > 0]

class Zaokraglenie:
    def __init__(self):
        self.name = "Zaokrąglenie do liczb całkowitych..."

    def wykonaj(self, data):
        print(self.name)
        return [round(x) for x in data]

class Skalowanie:
    def __init__(self):
        self.name = "Skalowanie..."

    def wykonaj(self, data):
        print(self.name)
        return [x*10 for x in data]

class Sumowanie:
    def __init__(self):
        self.name = "Sumowanie..."

    def wykonaj(self, data):
        print(self.name)
        return sum(data)

class Pipeline:
    def __init__(self, name):
        """
        Pipeline składa się z listy kroków (komponentów).
        Każdy komponent ma metodę wykonaj().
        """
        self.name = name
        self.kroki = []  # Lista komponentów

    def dodaj_krok(self, krok):
        """Dodaje komponent do pipeline'a"""
        self.kroki.append(krok)

    def uruchom(self, dane):
        """Wykonuje wszystkie kroki sekwencyjnie"""
        print(f"=== Pipeline: {self.name} ===\n")
        wynik = dane

        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
            print('Wynik operacji:')
            print(f'{wynik}\n')

        print(f"Wynik końcowy: {wynik}")
        return wynik

# Pipeline preparation
pipeline = Pipeline("Przetwarzanie danych liczbowych")
pipeline.dodaj_krok(FiltrowanieDodatnich())
pipeline.dodaj_krok(Zaokraglenie())
pipeline.dodaj_krok(Skalowanie())
pipeline.dodaj_krok(Sumowanie())

# Result
print(f'Dane wejściowe:\n{data}\n')
wynik = pipeline.uruchom(data)
```

# Zadanie 10 – Klasa Zamowienie z kompozycją

Stwórz klasę  Produkt  (nazwa, cena) oraz klasę  Zamowienie , która zawiera listę produktów.

Metody  Zamowienie:
- dodaj_produkt(produkt, ilosc)
- laczna_wartosc()  – suma wartości zamówienia
- pokaz_zamowienie()  – wyświetla wszystkie produkty i łączną wartość

Stwórz zamówienie z 5 produktami.

```python
class Produkt:
    def __init__(self, nazwa, cena):
        self.nazwa = nazwa
        self.cena = cena


class Zamowienie:
    def __init__(self):
        # lista elementów: (produkt, ilość)
        self.pozycje = []

    def dodaj_produkt(self, produkt, ilosc):
        self.pozycje.append((produkt, ilosc))

    def laczna_wartosc(self):
        return sum(produkt.cena * ilosc for produkt, ilosc in self.pozycje)

    def pokaz_zamowienie(self):
        print("Zamówienie:")
        for produkt, ilosc in self.pozycje:
            wartosc = produkt.cena * ilosc
            print(f"- {produkt.nazwa} x{ilosc} = {wartosc:.2f} zł")
        
        print(f"\nŁączna wartość: {self.laczna_wartosc():.2f} zł")

# Produkty
p1 = Produkt("Laptop", 3000)
p2 = Produkt("Mysz", 100)
p3 = Produkt("Klawiatura", 200)
p4 = Produkt("Monitor", 1200)
p5 = Produkt("Słuchawki", 300)

# Zamówienie
zamowienie = Zamowienie()

zamowienie.dodaj_produkt(p1, 1)
zamowienie.dodaj_produkt(p2, 2)
zamowienie.dodaj_produkt(p3, 1)
zamowienie.dodaj_produkt(p4, 2)
zamowienie.dodaj_produkt(p5, 3)

zamowienie.pokaz_zamowienie()
```

# Zadanie 12 – Dziedziczenie – FiguraGeometryczna

Stwórz klasę bazową  Figura  z metodą  pole().

Stwórz klasy potomne:
- Prostokat(szerokosc, wysokosc)
- Kolo(promien)

Każda klasa implementuje  pole().
Stwórz listę figur i oblicz łączne pole.

```python
import math
from abc import ABC, abstractmethod

# Klasy dziedziczące po klasie Figura muszą zawierać te same metody
# @abstractmethod zapewnia koltrolę sprawdzenia czy ten warunek jest spełniony już na etapie tworzenia kodu

class Figura(ABC):
    @abstractmethod
    def pole(self):
        raise NotImplementedError("Metoda pole() musi być zaimplementowana w klasie potomnej")

# klasy potomne
class Prostokat(Figura):
    def __init__(self, szerokosc, wysokosc):
        self.szerokosc = szerokosc
        self.wysokosc = wysokosc

    def pole(self):
        return self.szerokosc * self.wysokosc

class Kolo(Figura):
    def __init__(self, promien):
        self.promien = promien

    def pole(self):
        return math.pi * self.promien ** 2

# lista figur
figury = [
    Prostokat(4, 5),
    Kolo(3),
    Prostokat(2, 6),
    Kolo(1.5)
]

# łączne pole
laczne_pole = sum(figura.pole() for figura in figury)

print("Łączne pole:", laczne_pole)
```
