# Zadanie 5 – Parsowanie CSV

Masz string CSV:  "Alice,28,Data Scientist,75000". 

Użyj  split(',')  aby wyciągnąć: Imię, Wiek (jako int), Zawód, Pensję (jako int)

Wyświetl te informacje w formacie:  "Alice ma 28 lat, pracuje jako Data Scientist i zarabia $75000"

```python
# Data
csv_data = "Alice,28,Data Scientist,75000"

# Split 
fields = csv_data.split(',')
name, age, job, salary = fields

# Result
print(f'{name} ma {age} lat, pracuje jako {job} i zarabia ${salary}.')
```

# Zadanie 8 – Ekstakcja informacji ze stringa

Masz nazwę pliku:  "model_results_2024_Q3.csv". 

Wyciągnij rok i kwartał używając slicing lub split.

```python
# Data
file = "model_results_2024_Q3.csv"
fileformat = file.split('.')[-1] 
filename = file.strip(fileformat).strip('.')

# Details
year = filename[-7:-3]
quarter = filename[-2:]

# Result
print(f'Rok: {year}')
print(f'Kwartał: {quarter}')
```

# Zadanie 9 – Filtrowanie outlierów

Masz listę pomiarów:  [23, 24, 22, 150, 23, 25, -10, 24, 23]. 

Outlierem jest wartość poza zakresem [20, 30]. 

Stwórz nową listę bez outlierów używając list comprehension.

```python
# Data
measurements = [23, 24, 22, 150, 23, 25, -10, 24, 23]
boundries = [20, 30]

# Check
useful_measurements = [x for x in measurements if x >= boundries[0] and x <= boundries[1]]

# Result
print(useful_measurements)
```

#  Zadanie 10 – Agregacja danych po kategoriach

Masz listę transakcji (kategoria, kwota):

transactions = [
    ('food', 50),
    ('transport', 20),
    ('food', 30),
    ('entertainment', 40),
    ('transport', 15),
    ('food', 25)
]

Oblicz łączną kwotę dla każdej kategorii. Wynik zapisz w słowniku (użyj pętli i warunków).

```python
# Data
transactions = [
    ('food', 50),
    ('transport', 20),
    ('food', 30),
    ('entertainment', 40),
    ('transport', 15),
    ('food', 25)
]

# Calculation
unique_transactions = []
transaction_cost = []

for x in transactions:
    if x[0] not in unique_transactions:
        unique_transactions.append(x[0])
        transaction_cost.append(x[1])
    else:
        transaction_cost[unique_transactions.index(x[0])] += x[1]    
        
transaction_summary = dict(zip(unique_transactions, transaction_cost))

# Result
print(unique_transactions)
print(transaction_cost)
print(transaction_summary)
```
