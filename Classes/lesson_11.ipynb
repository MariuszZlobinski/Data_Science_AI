```python
# Przykład 1: Tworzenie Series i DataFrame - Podstawy

import pandas as pd
import numpy as np

# Series - jednowymiarowa struktura (jak kolumna)
temperatures = pd.Series([22.5, 23.1, 21.8, 24.2, 22.9],
                        index=['Poniedziałek', 'Wtorek', 'Środa', 'Czwartek', 'Piątek'])
print("Series - Temperatury:")
print(temperatures)
print(f"\nŚrednia temperatura: {temperatures.mean():.2f}°C")

# DataFrame - dwuwymiarowa struktura (jak tabela)
data = {
    'Produkt': ['Laptop', 'Mysz', 'Klawiatura', 'Monitor', 'Słuchawki'],
    'Cena': [3500, 80, 250, 1200, 350],
    'Sprzedane': [45, 320, 150, 78, 210],
    'Ocena': [4.5, 4.2, 4.7, 4.4, 4.3]
}

df = pd.DataFrame(data)
print("\nDataFrame - Produkty:")
print(df)

# Podstawowe informacje
print(f"\nKształt: {df.shape}")  # (wiersze, kolumny)
print(f"Kolumny: {df.columns.tolist()}")
print(f"Typy danych:\n{df.dtypes}")
```

```python
# Przykład 2: Wczytywanie z opcjami - Kontrola importu

import pandas as pd
import io

# Symulacja pliku CSV z niestandardowym separatorem i nagłówkiem
csv_data = """ID;Imię;Wiek;Miasto;Pensja
1;Anna Kowalska;28;Warszawa;7500
2;Jan Nowak;35;Kraków;8200
3;Maria Wiśniewska;42;Gdańsk;9100
4;Piotr Kamiński;31;Wrocław;7800
5;Zofia Lewandowska;29;Poznań;7600"""

# Wczytanie z niestandardowym separatorem
df_custom = pd.read_csv(
    io.StringIO(csv_data),  # Symulacja pliku
    sep=';',                 # Separator średnik
    encoding='utf-8',        # Kodowanie
    index_col='ID'          # Kolumna ID jako indeks
)

print("Dane z niestandardowym separatorem:")
print(df_custom)

# Wczytanie tylko wybranych kolumn
df_selected = pd.read_csv(
    io.StringIO(csv_data),
    sep=';',
    usecols=['Imię', 'Wiek', 'Pensja']  # Tylko te kolumny
)

print("\nTylko wybrane kolumny:")
print(df_selected)

# Wczytanie z pominięciem wierszy i limitowaniem
df_limited = pd.read_csv(io.StringIO(csv_data),
    sep=';',
    skiprows=[0],  # Pomijamy nagłówek (dla przykładu, tutaj go odczytamy normalnie)
    nrows=3        # Tylko pierwsze 3 wiersze danych
    )

print("\nOgraniczone do 3 wierszy:")
print(df_limited.head(5))
```

```python
# Przykład 3: Wczytywanie w batchach - Green IT approach

import pandas as pd

# Symulacja dużego pliku CSV
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# Wczytywanie w batchach (po 100 wierszy)
batch_size = 100
chunks = []

print("Wczytywanie w batchach:")
for i, chunk in enumerate(pd.read_csv(url, chunksize=batch_size)):
    print(f"Batch {i+1}: {len(chunk)} wierszy")
    chunks.append(chunk)

    # Przykład: przetwarzanie każdego batcha osobno
    print(f"  Średni wiek w batch {i+1}: {chunk['Age'].mean():.2f}")

    if i >= 2:  # Dla przykładu, tylko 3 batche
        break

# Połączenie wszystkich batchy (jeśli potrzebne)
df_combined = pd.concat(chunks, ignore_index=True)
print(f"\nPołączono {len(chunks)} batchy: {len(df_combined)} wierszy total")

# Alternatywa: przetwarzanie bez ładowania wszystkiego do pamięci
print("\nPrzetwarzanie bez ładowania całości:")
total_rows = 0
age_sum = 0
age_count = 0

for chunk in pd.read_csv(url, chunksize=100):
    total_rows += len(chunk)
    age_sum += chunk['Age'].sum()
    age_count += chunk['Age'].count()

print(f"Przetworzono {total_rows} wierszy")
print(f"Średni wiek (wszystkie dane): {age_sum / age_count:.2f}")
```

```python
# Wizualizacja: Porównanie metod wczytywania

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Symulacja czasów wczytywania i zużycia pamięci
methods = ['Pełny plik\n(default)', 'Wybrane kolumny\n(usecols)', 'Batch processing\n(chunksize)', 'Limit wierszy\n(nrows)']
time = [100, 60, 45, 25]  # Względny czas (%)
memory = [100, 40, 20, 30]  # Względne zużycie pamięci (%)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Wykres czasu
ax1.barh(methods, time, color=['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4'])
ax1.set_xlabel('Względny czas wczytywania (%)', fontsize=11)
ax1.set_title('Porównanie czasu wczytywania danych', fontsize=13, 
fontweight='bold')

ax1.set_xlim(0, 110)
for i, v in enumerate(time):
    ax1.text(v + 2, i, f'{v}%', va='center', fontsize=10)

# Wykres pamięci
ax2.barh(methods, memory, color=['#ff6b6b', '#4ecdc4', '#45b7d1', 
'#96ceb4'])
ax2.set_xlabel('Względne zużycie pamięci (%)', fontsize=11)
ax2.set_title('Porównanie zużycia pamięci', fontsize=13, fontweight='bold')
ax2.set_xlim(0, 110)
for i, v in enumerate(memory):
    ax2.text(v + 2, i, f'{v}%', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('pandas_loading_comparison.png', dpi=100, bbox_inches='tight')
print("Wykres zapisany jako: pandas_loading_comparison.png")
# plt.show()  # Odkomentuj aby wyświetlić
```

```python
# Przykład 2: Filtrowanie warunkowe - Boolean indexing

import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Filtr: pasażerowie którzy przeżyli
survived = titanic[titanic['Survived'] == 1]
print(f"Liczba ocalonych: {len(survived)}")
print(survived.head())

# Filtr: kobiety które przeżyły
survived_women = titanic[(titanic['Survived'] == 1) & (titanic['Sex'] == 
'female')]
print(f"\nLiczba ocalonych kobiet: {len(survived_women)}")

# Filtr: pasażerowie w wieku 20-40 lat
age_20_40 = titanic[(titanic['Age'] >= 20) & (titanic['Age'] <= 40)]
print(f"\nPasażerowie 20-40 lat: {len(age_20_40)}")

# Filtr: pierwsza lub druga klasa
first_second_class = titanic[titanic['Pclass'].isin([1, 2])]
print(f"\nPasażerowie 1. lub 2. klasy: {len(first_second_class)}")

# Filtr: wykluczenie wartości
not_third_class = titanic[~titanic['Pclass'].isin([3])]  # ~ to negacja
print(f"Pasażerowie NIE z 3. klasy: {len(not_third_class)}")

# Złożony filtr
complex_filter = titanic[
    (titanic['Age'] > 30) &
    (titanic['Sex'] == 'male') &
    (titanic['Pclass'] == 1) &
    (titanic['Survived'] == 1)
]

print(f"\nOcaleni mężczyźni >30 lat z 1. klasy: {len(complex_filter)}")
print(complex_filter[['Name', 'Age', 'Pclass', 'Survived']].head())
```

```python
# Przykład 3: Metody .query() i .filter() - Alternatywna składnia

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Metoda .query() - bardziej czytelna składnia dla złożonych warunków
print("Filtrowanie przez .query():")
result = titanic.query("Age > 30 and Sex == 'female' and Survived == 1")
print(f"Ocalałe kobiety >30 lat: {len(result)}")
print(result[['Name', 'Age', 'Sex', 'Survived']].head())

# .query() z zmiennymi
min_age = 25
max_age = 35
result2 = titanic.query("Age >= @min_age and Age <= @max_age")
print(f"\nPasażerowie {min_age}-{max_age} lat: {len(result2)}")

# Metoda .filter() - filtrowanie KOLUMN po nazwach
print("\n.filter() - kolumny zawierające 'S':")
filtered_cols = titanic.filter(like='S')  # Kolumny z literą 'S'
print(filtered_cols.columns.tolist())
print(filtered_cols.head())

# .filter() z regex
print("\n.filter() - kolumny zaczynające się od 'S' lub 'P':")
regex_cols = titanic.filter(regex='^(S|P)')
print(regex_cols.columns.tolist())
```

```python
# Wizualizacja: Proporcje przeżycia

import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Analiza przeżycia wg klasy i płci
survival_data = titanic.groupby(['Pclass', 'Sex'])['Survived'].agg(['sum', 'count'])
survival_data['survival_rate'] = (survival_data['sum'] / survival_data['count']) * 100
print("Wskaźniki przeżycia:")
print(survival_data)

# Przygotowanie danych do wykresu
survival_pivot = titanic.pivot_table(
    values='Survived',
    index='Pclass',
    columns='Sex',
    aggfunc='mean'
) * 100

# Wykres słupkowy
fig, ax = plt.subplots(figsize=(10, 6))
survival_pivot.plot(kind='bar', ax=ax, color=['#ff6b6b', '#4ecdc4'], width=0.7)

ax.set_xlabel('Klasa pasażera', fontsize=12, fontweight='bold')
ax.set_ylabel('Wskaźnik przeżycia (%)', fontsize=12, fontweight='bold')
ax.set_title('Wskaźnik przeżycia na Titanicu wg klasy i płci', fontsize=14, 
fontweight='bold')
ax.set_xticklabels(['1. klasa', '2. klasa', '3. klasa'], rotation=0)
ax.legend(title='Płeć', labels=['Kobiety', 'Mężczyźni'])
ax.set_ylim(0, 100)
ax.grid(axis='y', alpha=0.3)

# Dodanie wartości na słupkach
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=3)
    
plt.tight_layout()
plt.savefig('titanic_survival_rate.png', dpi=100, bbox_inches='tight')
print("\nWykres zapisany jako: titanic_survival_rate.png")
# plt.show()
```

# 4. Obsługa brakujących danych

```python
# Przykład 1: Wykrywanie brakujących danych

import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Sprawdzenie brakujących wartości
print("Liczba NaN w każdej kolumnie:")
print(titanic.isnull().sum())  # lub .isna()

# Procent brakujących wartości
print("\nProcent brakujących wartości:")
missing_percent = (titanic.isnull().sum() / len(titanic)) * 100
print(missing_percent[missing_percent > 0])

# Liczba wierszy z JAKIMIKOLWIEK brakującymi wartościami
rows_with_nan = titanic.isnull().any(axis=1).sum()
print(f"\nWiersze z co najmniej 1 NaN: {rows_with_nan} ({rows_with_nan/len(titanic)*100:.1f}%)")

# Wizualizacja brakujących danych - mapa ciepła
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
missing_data = titanic.isnull().astype(int)
ax.imshow(missing_data.T, cmap='RdYlGn_r', aspect='auto', 
interpolation='none')
ax.set_yticks(range(len(titanic.columns)))
ax.set_yticklabels(titanic.columns)
ax.set_xlabel('Indeks wiersza', fontsize=11)
ax.set_title('Mapa brakujących danych (czerwony = brak)', fontsize=13, 
fontweight='bold')
plt.tight_layout()
plt.savefig('missing_data_heatmap.png', dpi=100, bbox_inches='tight')
print("\nMapa zapisana jako: missing_data_heatmap.png")
# plt.show()
```

```python
# Przykład 2: Usuwanie brakujących danych - dropna()

import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

print(f"Oryginalny kształt: {titanic.shape}")

# Usunięcie wierszy z JAKIMIKOLWIEK brakującymi wartościami
titanic_dropped_any = titanic.dropna()
print(f"Po dropna() (any): {titanic_dropped_any.shape}")  # Bardzo mało wierszy!

# Usunięcie wierszy gdzie WSZYSTKIE wartości to NaN
titanic_dropped_all = titanic.dropna(how='all')
print(f"Po dropna(how='all'): {titanic_dropped_all.shape}")  # Wszystkie wiersze (brak takich)

# Usunięcie wierszy z NaN w konkretnych kolumnach
titanic_dropped_subset = titanic.dropna(subset=['Age', 'Embarked'])
print(f"Po dropna(subset=['Age', 'Embarked']): {titanic_dropped_subset.shape}")

# Usunięcie KOLUMN z dużą ilością NaN (>50%)
threshold = len(titanic) * 0.5  # 50% wierszy
titanic_dropped_cols = titanic.dropna(thresh=threshold, axis=1)
print(f"Po usunięciu kolumn z >50% NaN: {titanic_dropped_cols.shape}")
print(f"Usunięte kolumny: {set(titanic.columns) - set(titanic_dropped_cols.columns)}")

# Statystyki przed i po
print("\n--- Statystyki Age ---")
print(f"Przed: średnia = {titanic['Age'].mean():.2f}, count = {titanic['Age'].count()}")
print(f"Po: średnia = {titanic_dropped_subset['Age'].mean():.2f}, count = {titanic_dropped_subset['Age'].count()}")
```

```python
# Przykład 3: Wypełnianie brakujących danych - fillna()

import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Kopia do testów
titanic_filled = titanic.copy()

# Strategia 1: Wypełnienie konkretną wartością
titanic_filled['Age'].fillna(0, inplace=True)
print("Age po wypełnieniu zerami (pierwsze 10 oryginalnych NaN):")
print(titanic_filled.loc[titanic['Age'].isnull(), 'Age'].head(10))

# Strategia 2: Wypełnienie średnią (dla danych numerycznych)
titanic_filled = titanic.copy()
age_mean = titanic['Age'].mean()
titanic_filled['Age'].fillna(age_mean, inplace=True)
print(f"\nAge po wypełnieniu średnią ({age_mean:.2f}):")
print(titanic_filled['Age'].describe())

# Strategia 3: Wypełnienie medianą (bardziej odporne na outliers)
titanic_filled = titanic.copy()
age_median = titanic['Age'].median()
titanic_filled['Age'].fillna(age_median, inplace=True)
print(f"\nAge po wypełnieniu medianą ({age_median:.2f}):")
print(titanic_filled['Age'].describe())

# Strategia 4: Wypełnienie dominantą (dla kategorii)
titanic_filled = titanic.copy()
embarked_mode = titanic['Embarked'].mode()[0]  # Najczęstsza wartość
titanic_filled['Embarked'].fillna(embarked_mode, inplace=True)
print(f"\nEmbarked po wypełnieniu dominantą ('{embarked_mode}'):")
print(titanic_filled['Embarked'].value_counts())

# Strategia 5: Forward fill (propagacja wartości)
titanic_filled = titanic.copy()
titanic_filled['Age'] = titanic_filled['Age'].ffill()  # Poprzednia wartość
print("\nAge po forward fill (pierwsze 20):")
print(titanic_filled['Age'].head(20))

# Strategia 6: Backward fill
titanic_filled = titanic.copy()
titanic_filled['Age'] = titanic_filled['Age'].bfill()  # Następna wartość
print("\nAge po backward fill (pierwsze 20):")
print(titanic_filled['Age'].head(20))

# Porównanie strategii
print("\n--- Porównanie średniej Age ---")
print(f"Oryginał (bez NaN): {titanic['Age'].mean():.2f}")
print(f"Wypełnienie 0: różnica dużo!")
print(f"Wypełnienie średnią: {age_mean:.2f} (zachowana średnia)")
print(f"Wypełnienie medianą: {age_median:.2f}")
```

```python
# Przykład 4: Zaawansowane techniki - interpolacja

import pandas as pd
import numpy as np

# Stworzenie przykładowych szeregów czasowych z lukami
dates = pd.date_range('2024-01-01', periods=20, freq='D')
values = [10, 12, np.nan, np.nan, 18, 20, np.nan, 25, 27, np.nan,
          np.nan, np.nan, 35, 38, 40, np.nan, 45, 47, 50, np.nan]

df_time = pd.DataFrame({'data': dates, 'wartość': values})
df_time.set_index('data', inplace=True)

print("Dane z lukami:")
print(df_time)

# Interpolacja liniowa
df_linear = df_time.copy()
df_linear['wartość'] = df_linear['wartość'].interpolate(method='linear')
print("\nPo interpolacji liniowej:")
print(df_linear)

# Interpolacja czasowa (uwzględnia nierówne odstępy czasowe)
df_time_interp = df_time.copy()
df_time_interp['wartość'] = df_time_interp['wartość'].interpolate(method='time')
print("\nPo interpolacji czasowej:")
print(df_time_interp)

# Wizualizacja
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(df_time.index, df_time['wartość'], 'o--', label='Oryginał (z NaN)',
        markersize=8, color='red', alpha=0.6)
ax.plot(df_linear.index, df_linear['wartość'], 's-', label='Interpolacja liniowa',
        markersize=6, color='blue', alpha=0.8)
ax.set_xlabel('Data', fontsize=11)
ax.set_ylabel('Wartość', fontsize=11)
ax.set_title('Interpolacja brakujących danych w szeregu czasowym', 
fontsize=13, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('interpolation_example.png', dpi=100, bbox_inches='tight')
print("\nWykres zapisany jako: interpolation_example.png")
# plt.show()
```

```python
# 2. Optymalizacja typów danych - redukcja zużycia pamięci

import pandas as pd
import numpy as np

# Wczytanie danych
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

print(f"Zużycie pamięci przed optymalizacją: {df.memory_usage(deep=True).sum() / 1024:.2f} KB")

# Optymalizacja typów
df_optimized = df.copy()

# int64 → int8/int16 (gdzie możliwe)
df_optimized['Pclass'] = df_optimized['Pclass'].astype('int8')  # 1-3 mieści się w int8
df_optimized['Survived'] = df_optimized['Survived'].astype('int8')  # 0-1
df_optimized['SibSp'] = df_optimized['SibSp'].astype('int8')
df_optimized['Parch'] = df_optimized['Parch'].astype('int8')

# float64 → float32 (jeśli precyzja wystarczająca)
df_optimized['Age'] = df_optimized['Age'].astype('float32')
df_optimized['Fare'] = df_optimized['Fare'].astype('float32')

# object → category (dla kolumn z małą liczbą unikalnych wartości)
df_optimized['Sex'] = df_optimized['Sex'].astype('category')
df_optimized['Embarked'] = df_optimized['Embarked'].astype('category')
print(f"Zużycie pamięci po optymalizacji: {df_optimized.memory_usage(deep=True).sum() / 1024:.2f} KB")
reduction = (1 - df_optimized.memory_usage(deep=True).sum() / 
df.memory_usage(deep=True).sum()) * 100
print(f"Redukcja: {reduction:.1f}%")

# Korzyści Green IT:
# 🟢 Mniej pamięci = mniej energii zużywanej przez RAM
# 🟢 Szybsze operacje (cache efficiency)
# 🟢 Możliwość przetwarzania większych zbiorów
```

```python
# 4. Monitoring zużycia zasobów

import pandas as pd
import psutil
import os

def get_memory_usage():
    """Zwraca zużycie pamięci przez proces w MB"""
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / 1024 / 1024

# Przed operacją
mem_before = get_memory_usage()

# Operacja
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Po operacji
mem_after = get_memory_usage()

print(f"Zużycie pamięci przed: {mem_before:.2f} MB")
print(f"Zużycie pamięci po: {mem_after:.2f} MB")
print(f"Wzrost: {mem_after - mem_before:.2f} MB")
```

```python
# Dataset: Titanic - Czyszczenie i przygotowanie

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

print("=" * 50)
print("KROK 1: PIERWSZE SPOJRZENIE NA DANE")
print("=" * 50)
print(titanic.head())
print(f"\nKształt: {titanic.shape}")
print(f"\nTypy danych:\n{titanic.dtypes}")

print("\n" + "=" * 50)
print("KROK 2: ANALIZA BRAKUJĄCYCH DANYCH")
print("=" * 50)
missing = titanic.isnull().sum()
missing_percent = (missing / len(titanic)) * 100
missing_df = pd.DataFrame({
    'Kolumna': missing.index,
    'Brakujące': missing.values,
    'Procent': missing_percent.values
})
print(missing_df[missing_df['Brakujące'] > 0])

print("\n" + "=" * 50)
print("KROK 3: CZYSZCZENIE DANYCH")
print("=" * 50)

# Kopia do czyszczenia
titanic_clean = titanic.copy()

# 1. Usunięcie kolumny 'Cabin' (77% NaN - zbyt dużo braków)
titanic_clean.drop('Cabin', axis=1, inplace=True)
print("✓ Usunięto kolumnę 'Cabin' (77% NaN)")

# 2. Wypełnienie 'Age' medianą (20% NaN)
age_median = titanic_clean['Age'].median()
titanic_clean['Age'].fillna(age_median, inplace=True)
print(f"✓ Wypełniono 'Age' medianą: {age_median}")

# 3. Wypełnienie 'Embarked' dominantą (2 NaN)
embarked_mode = titanic_clean['Embarked'].mode()[0]
titanic_clean['Embarked'].fillna(embarked_mode, inplace=True)
print(f"✓ Wypełniono 'Embarked' dominantą: '{embarked_mode}'")

# 4. Weryfikacja
print(f"\nBrakujące wartości po czyszczeniu:")
print(titanic_clean.isnull().sum().sum())  # Powinno być 0

# 5. Optymalizacja typów
titanic_clean['Pclass'] = titanic_clean['Pclass'].astype('int8')
titanic_clean['Survived'] = titanic_clean['Survived'].astype('int8')
titanic_clean['SibSp'] = titanic_clean['SibSp'].astype('int8')
titanic_clean['Parch'] = titanic_clean['Parch'].astype('int8')
titanic_clean['Sex'] = titanic_clean['Sex'].astype('category')
titanic_clean['Embarked'] = titanic_clean['Embarked'].astype('category')

print(f"✓ Zoptymalizowano typy danych")

print("\n" + "=" * 50)
print("KROK 4: STATYSTYKI OPISOWE")
print("=" * 50)
print(titanic_clean.describe())

print("\n" + "=" * 50)
print("KROK 5: ANALIZA KATEGORII")
print("=" * 50)
print("Rozkład płci:")
print(titanic_clean['Sex'].value_counts())
print("\nRozkład klas:")
print(titanic_clean['Pclass'].value_counts().sort_index())
print("\nRozkład portów:")
print(titanic_clean['Embarked'].value_counts())
```

```python
# Eksploracyjna Analiza Danych (EDA)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Kontynuacja z poprzedniego przykładu (titanic_clean)
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic_clean = titanic.copy()
titanic_clean.drop('Cabin', axis=1, inplace=True)
titanic_clean['Age'].fillna(titanic_clean['Age'].median(), inplace=True)
titanic_clean['Embarked'].fillna(titanic_clean['Embarked'].mode()[0], 
inplace=True)

# Wizualizacje
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Rozkład wieku
axes[0, 0].hist(titanic_clean['Age'], bins=30, color='skyblue', 
edgecolor='black', alpha=0.7)
axes[0, 0].axvline(titanic_clean['Age'].mean(), color='red', linestyle='--', 
label=f'Średnia: {titanic_clean["Age"].mean():.1f}')
axes[0, 0].axvline(titanic_clean['Age'].median(), color='green', 
linestyle='--', label=f'Mediana: {titanic_clean["Age"].median():.1f}')
axes[0, 0].set_xlabel('Wiek', fontsize=11)
axes[0, 0].set_ylabel('Liczba pasażerów', fontsize=11)
axes[0, 0].set_title('Rozkład wieku pasażerów', fontsize=12, 
fontweight='bold')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# 2. Przeżycie wg płci
survival_by_sex = titanic_clean.groupby('Sex')['Survived'].mean() * 100
survival_by_sex.plot(kind='bar', ax=axes[0, 1], color=['#ff6b6b', 
'#4ecdc4'], alpha=0.8)
axes[0, 1].set_ylabel('Wskaźnik przeżycia (%)', fontsize=11)
axes[0, 1].set_xlabel('Płeć', fontsize=11)
axes[0, 1].set_title('Wskaźnik przeżycia wg płci', fontsize=12, 
fontweight='bold')
axes[0, 1].set_xticklabels(['Kobiety', 'Mężczyźni'], rotation=0)
axes[0, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(survival_by_sex.values):
    axes[0, 1].text(i, v + 2, f'{v:.1f}%', ha='center', fontsize=10, fontweight='bold')

# 3. Rozkład cen biletów (log scale)
axes[1, 0].hist(titanic_clean['Fare'], bins=50, color='coral', 
edgecolor='black', alpha=0.7)
axes[1, 0].set_xlabel('Cena biletu (£)', fontsize=11)
axes[1, 0].set_ylabel('Liczba pasażerów', fontsize=11)
axes[1, 0].set_title('Rozkład cen biletów', fontsize=12, fontweight='bold')
axes[1, 0].set_yscale('log')  # Log scale bo długi ogon
axes[1, 0].grid(alpha=0.3)

# 4. Przeżycie wg klasy
survival_by_class = titanic_clean.groupby('Pclass')['Survived'].mean() * 100
survival_by_class.plot(kind='bar', ax=axes[1, 1], color=['gold', 'silver', 
'#cd7f32'], alpha=0.8)
axes[1, 1].set_ylabel('Wskaźnik przeżycia (%)', fontsize=11)
axes[1, 1].set_xlabel('Klasa', fontsize=11)
axes[1, 1].set_title('Wskaźnik przeżycia wg klasy pasażera', fontsize=12, 
fontweight='bold')
axes[1, 1].set_xticklabels(['1. klasa', '2. klasa', '3. klasa'], rotation=0)
axes[1, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(survival_by_class.values):
    axes[1, 1].text(i, v + 2, f'{v:.1f}%', ha='center', fontsize=10, fontweight='bold')
    
plt.tight_layout()
plt.savefig('titanic_eda.png', dpi=100, bbox_inches='tight')
print("Wykres EDA zapisany jako: titanic_eda.png")
# plt.show()
```

```python
# Feature Engineering - Tworzenie nowych zmiennych

import pandas as pd
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic_clean = titanic.copy()
titanic_clean.drop('Cabin', axis=1, inplace=True)
titanic_clean['Age'].fillna(titanic_clean['Age'].median(), inplace=True)
titanic_clean['Embarked'].fillna(titanic_clean['Embarked'].mode()[0], 
inplace=True)

# Feature 1: Rozmiar rodziny
titanic_clean['FamilySize'] = titanic_clean['SibSp'] + titanic_clean['Parch'] + 1
print("Rozmiar rodziny:")
print(titanic_clean['FamilySize'].value_counts().head())

# Feature 2: Czy pasażer był sam?
titanic_clean['IsAlone'] = (titanic_clean['FamilySize'] == 1).astype(int)
print(f"\nPasażerowie samotni: {titanic_clean['IsAlone'].sum()}")

# Feature 3: Kategoria wieku
bins = [0, 12, 18, 35, 60, 100]
labels = ['Dziecko', 'Nastolatek', 'Dorosły', 'Starszy', 'Senior']
titanic_clean['AgeGroup'] = pd.cut(titanic_clean['Age'], bins=bins, 
labels=labels)
print("\nGrupy wiekowe:")
print(titanic_clean['AgeGroup'].value_counts())

# Feature 4: Kategoria ceny biletu
fare_bins = [0, 10, 30, 100, 600]
fare_labels = ['Tani', 'Średni', 'Drogi', 'Bardzo drogi']
titanic_clean['FareCategory'] = pd.cut(titanic_clean['Fare'], 
bins=fare_bins, labels=fare_labels)
print("\nKategorie cen:")
print(titanic_clean['FareCategory'].value_counts())
# Wyświetlenie przykładów
print("\nPrzykładowe wiersze z nowymi zmiennymi:")
print(titanic_clean[['Name', 'Age', 'AgeGroup', 'FamilySize', 'IsAlone', 'Fare', 'FareCategory']].head(10))
```
