# Zadanie 10 – Custom Funkcja Agregująca

Napisz własną funkcję, która oblicza zakres międzykwartylowy (IQR = Q3 - Q1) i zastosuj ją do grupy.

Dataset: California Housing

Wymagania:
- Napisz funkcję  iqr_func(x)  obliczającą IQR
- Zastosuj ją do  MedHouseVal  według regionów
- Porównaj IQR między regionami

Oczekiwany rezultat:
- Tabela z IQR dla każdego regionu

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Data
housing = fetch_california_housing()

df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
df_housing = df.copy()

# Region specification
lat_bins = pd.cut(df['Latitude'], bins=2, labels=['South', 'North'])
lon_bins = pd.cut(df['Longitude'], bins=2, labels=['West', 'East'])

# Add 'Region' column
df['Region'] = lat_bins.astype(str) + "_" + lon_bins.astype(str)

# Quantile method
def iqr_func(x):
    return x.quantile(0.75) - x.quantile(0.25)

# Result
iqr_by_region = df.groupby('Region')['MedHouseVal'].agg(iqr_func)

print("IQR MedHouseVal dla regionów:")
print(iqr_by_region)
```

```python
# Just to compare
print("\n\nLiczba obserwacji według regionu (.cut())")
print(df['Region'].value_counts())

df_housing['Region'] = 'Unknown'
df_housing.loc[(df_housing['Latitude'] >= 36) & 
               (df_housing['Longitude'] >= -119), 'Region'] = 'NE'  # Północny Wschód
df_housing.loc[(df_housing['Latitude'] >= 36) & 
               (df_housing['Longitude'] < -119), 'Region'] = 'NW'   # Północny Zachód
df_housing.loc[(df_housing['Latitude'] < 36) & 
               (df_housing['Longitude'] >= -119), 'Region'] = 'SE'    # Południowy Wschód
df_housing.loc[(df_housing['Latitude'] < 36) & 
               (df_housing['Longitude'] < -119), 'Region'] = 'SW'     # Południowy Zachód
print("\nLiczba obserwacji według regionu (ręczny podział)")
print(df_housing['Region'].value_counts())

iqr_by_region2 = df_housing.groupby('Region')['MedHouseVal'].agg(iqr_func)
print("\n\nIQR MedHouseVal dla regionów: (ręczny podział)")
print(iqr_by_region2)



def iqr_func2(x):
    return x.quantile(0.75)
def iqr_func3(x):
    return x.quantile(0.25)

print(f'\n\n{df.groupby('Region')['MedHouseVal'].agg(iqr_func2)}')
print(df.groupby('Region')['MedHouseVal'].agg(iqr_func3))

```
