# 1. Wprowadzenie do OOP – klasy i obiekty

```python
# Przykład 1: Podstawowa klasa – tworzenie obiektu

# Prosta klasa reprezentująca dataset
class Dataset:
    def __init__(self, nazwa):
        """
        Konstruktor klasy Dataset.
        __init__ to specjalna metoda wywoływana podczas tworzenia obiektu.
        'self' to referencja do aktualnego obiektu.
        """
        self.nazwa = nazwa  # Atrybut (cecha) obiektu
        self.dane = []      # Lista na dane

    def dodaj_wiersz(self, wiersz):
        """Metoda dodająca wiersz do datasetu"""
        self.dane.append(wiersz)

    def pokaz_info(self):
        """Metoda wyświetlająca informacje o datasecie"""
        print(f"Dataset: {self.nazwa}")
        print(f"Liczba wierszy: {len(self.dane)}")

# Tworzenie obiektów (instancji klasy)
ds1 = Dataset("Sprzedaż Q1")
ds1.dodaj_wiersz([100, 200, 300])
ds1.dodaj_wiersz([150, 250, 350])
ds1.pokaz_info()

# Drugi obiekt tej samej klasy
ds2 = Dataset("Sprzedaż Q2")
ds2.dodaj_wiersz([180, 220, 320])
ds2.pokaz_info()
```

```python
# Przykład 2: Klasa z metodami obliczeniowymi

# Klasa reprezentująca dane liczbowe z funkcjami statystycznymi
class StatDataset:
    def __init__(self, nazwa, wartosci):
        """
        Klasa do przechowywania danych i obliczania podstawowych statystyk.
        """
        self.nazwa = nazwa
        self.wartosci = wartosci  # Lista liczb

    def srednia(self):
        """Oblicza średnią arytmetyczną"""
        return sum(self.wartosci) / len(self.wartosci)
    
    def minimum(self):
        """Zwraca wartość minimalną"""
        return min(self.wartosci)
    
    def maksimum(self):
        """Zwraca wartość maksymalną"""
        return max(self.wartosci)
    
    def raport(self):
        """Generuje pełny raport statystyczny"""
        print(f"=== Raport dla: {self.nazwa} ===")
        print(f"Średnia: {self.srednia():.2f}")
        print(f"Min: {self.minimum()}")
        print(f"Max: {self.maksimum()}")
        print(f"Zakres: {self.maksimum() - self.minimum()}")

# Użycie klasy
dane_sprzedazy = StatDataset("Sprzedaż miesięczna", [1200, 1500, 1350, 1800, 1600, 1700])
dane_sprzedazy.raport()

dane_temp = StatDataset("Temperatura dzienna", [18.5, 20.1, 19.8, 22.3, 
21.0, 20.5, 19.2])
dane_temp.raport()
```

```python
# Przykład 3: Klasa modelująca obserwacje Data Science

# Klasa reprezentująca pojedynczą obserwację w datasecie
class Obserwacja:
    def __init__(self, id, cechy, etykieta=None):
        """
        Pojedyncza obserwacja w datasecie ML.
        Parameters:
        - id: identyfikator obserwacji
        - cechy: lista lub słownik cech (features)
        - etykieta: etykieta (label) dla uczenia nadzorowanego (opcjonalna)
        """
        self.id = id
        self.cechy = cechy
        self.etykieta = etykieta

    def pokaz(self):
        """Wyświetla informacje o obserwacji"""
        print(f"ID: {self.id}")
        print(f"Cechy: {self.cechy}")
        if self.etykieta is not None:
            print(f"Etykieta: {self.etykieta}")
        else:
            print("Brak etykiety (dane unlabeled)")

    def ma_etykiete(self):
        """Sprawdza, czy obserwacja ma etykietę"""
        return self.etykieta is not None
    
# Tworzenie obserwacji
obs1 = Obserwacja(1, {"wiek": 25, "wzrost": 170, "waga": 65}, 
etykieta="zdrowy")
obs1.pokaz()

print(f"\nCzy ma etykietę: {obs1.ma_etykiete()}")

obs2 = Obserwacja(2, [1.2, 0.8, 2.3, 1.5])  # Obserwacja bez etykiety
obs2.pokaz()
print(f"Czy ma etykietę: {obs2.ma_etykiete()}")
    
    
```

# 2. Dziedziczenie – specjalizacja klas

```python
# Przykład 1: Dziedziczenie – klasa bazowa i potomna

# Klasa bazowa (rodzic)
class Model:
    def __init__(self, nazwa):
        """Bazowa klasa dla wszystkich modeli ML"""
        self.nazwa = nazwa
        self.wytrenowany = False

    def trenuj(self, dane):
        """Podstawowa metoda trenowania (placeholder)"""
        print(f"Trenuję model {self.nazwa}...")
        self.wytrenowany = True

    def czy_wytrenowany(self):
        """Sprawdza, czy model jest wytrenowany"""
        return self.wytrenowany
    
# Klasa potomna (dziecko) dziedziczy po Model
class ModelRegresji(Model):
    def __init__(self, nazwa, typ_regresji="liniowa"):
        """
        Konstruktor klasy potomnej.
        super().__init__(nazwa) wywołuje konstruktor klasy bazowej.
        """
        super().__init__(nazwa)  # Wywołanie konstruktora rodzica
        self.typ_regresji = typ_regresji  # Nowy atrybut specyficzny dla regresji

    def przewiduj(self, x):
        """Metoda specyficzna dla regresji"""
        if not self.wytrenowany:
            print("Błąd: Model nie jest wytrenowany!")
            return None
        # Prosta symulacja predykcji
        return 2.5 * x + 1.3
    
# Użycie
model_reg = ModelRegresji("Regresja Sprzedaży", typ_regresji="liniowa")
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")

model_reg.trenuj([1, 2, 3, 4, 5])  # Metoda z klasy bazowej
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")

predykcja = model_reg.przewiduj(10)  # Metoda specyficzna dla ModelRegresji
print(f"Predykcja dla x=10: {predykcja}")
```

```python
# Przykład 2: Hierarchia klas – wiele klas potomnych

# Klasa bazowa
class Transformer:
    def __init__(self, nazwa):
        """Bazowy transformer do przetwarzania danych"""
        self.nazwa = nazwa

    def transformuj(self, dane):
        """Metoda bazowa – do nadpisania w klasach potomnych"""
        raise NotImplementedError("Metoda transformuj() musi być zaimplementowana!")
    
# Klasa potomna 1: Normalizacja
class Normalizator(Transformer):
    def __init__(self):
        super().__init__("Normalizator")
        self.min_val = None
        self.max_val = None

    def transformuj(self, dane):
        """Normalizacja min-max do zakresu [0, 1]"""
        self.min_val = min(dane)
        self.max_val = max(dane)
        zakres = self.max_val - self.min_val
        if zakres == 0:
            return [0.5] * len(dane)  # Wszystkie wartości identyczne
        return [(x - self.min_val) / zakres for x in dane]
    
# Klasa potomna 2: Standaryzacja
class Standaryzator(Transformer):
    def __init__(self):
        super().__init__("Standaryzator")
        self.srednia = None
        self.odch_std = None

    def transformuj(self, dane):
        """Standaryzacja do rozkładu o średniej 0 i odchyleniu std 1"""
        self.srednia = sum(dane) / len(dane)
        wariancja = sum((x - self.srednia)**2 for x in dane) / len(dane)
        self.odch_std = wariancja ** 0.5

        if self.odch_std == 0:
            return [0.0] * len(dane)
        return [(x - self.srednia) / self.odch_std for x in dane]
    
# Użycie różnych transformerów
dane = [10, 20, 30, 40, 50]

norm = Normalizator()
dane_norm = norm.transformuj(dane)
print(f"Dane oryginalne: {dane}")
print(f"Po normalizacji: {[round(x, 2) for x in dane_norm]}")
print(f"Min: {norm.min_val}, Max: {norm.max_val}\n")

stand = Standaryzator()
dane_stand = stand.transformuj(dane)
print(f"Po standaryzacji: {[round(x, 2) for x in dane_stand]}")
print(f"Średnia: {stand.srednia}, Odch. std: {round(stand.odch_std, 2)}")
```

```python
# Przykład 3: Nadpisywanie metod (override)

# Klasa bazowa
class Raport:
    def __init__(self, tytul):
        self.tytul = tytul

    def generuj(self):
        """Generuje podstawowy raport"""
        print(f"=== {self.tytul} ===")
        print("Brak szczegółowych danych.")

# Klasa potomna nadpisuje metodę generuj()
class RaportSzczegolowy(Raport):
    def __init__(self, tytul, dane):
        super().__init__(tytul)
        self.dane = dane

    def generuj(self):
        """Nadpisana metoda – generuje szczegółowy raport"""
        print(f"=== {self.tytul} ===")
        print(f"Liczba obserwacji: {len(self.dane)}")
        print(f"Średnia: {sum(self.dane)/len(self.dane):.2f}")
        print(f"Min: {min(self.dane)}, Max: {max(self.dane)}")

# Użycie
raport1 = Raport("Raport podstawowy")
raport1.generuj()

print()

raport2 = RaportSzczegolowy("Raport sprzedaży", [100, 150, 200, 180, 220])
raport2.generuj()  # Wywołuje nadpisaną metodę
```

# 3. Kompozycja – budowanie klas z innych obiektów

```python
# Przykład 1: Kompozycja – pipeline przetwarzania danych

# Proste komponenty
class Czyszczenie:
    def __init__(self):
        self.nazwa = "Czyszczenie danych"

    def wykonaj(self, dane):
        """Usuwa wartości ujemne (symulacja czyszczenia)"""
        print(f"  [{self.nazwa}] Usuwam wartości ujemne...")
        return [x for x in dane if x >= 0]
    
class Normalizacja:
    def __init__(self):
        self.nazwa = "Normalizacja"

    def wykonaj(self, dane):
        """Normalizacja do zakresu [0, 1]"""
        print(f"  [{self.nazwa}] Normalizuję dane...")
        if len(dane) == 0:
            return []
        min_val = min(dane)
        max_val = max(dane)
        zakres = max_val - min_val
        if zakres == 0:
            return [0.5] * len(dane)
        return [(x - min_val) / zakres for x in dane]
    
class Agregacja:
    def __init__(self):
        self.nazwa = "Agregacja"

    def wykonaj(self, dane):
        """Oblicza średnią jako wynik agregacji"""
        print(f"  [{self.nazwa}] Obliczam średnią...")
        if len(dane) == 0:
            return 0
        return sum(dane) / len(dane)
    
# Klasa Pipeline używa kompozycji – zawiera inne obiekty jako komponenty
class Pipeline:
    def __init__(self, nazwa):
        """
        Pipeline składa się z listy kroków (komponentów).
        Każdy komponent ma metodę wykonaj().
        """
        self.nazwa = nazwa
        self.kroki = []  # Lista komponentów

    def dodaj_krok(self, krok):
        """Dodaje komponent do pipeline'a"""
        self.kroki.append(krok)

    def uruchom(self, dane):
        """Wykonuje wszystkie kroki sekwencyjnie"""
        print(f"=== Pipeline: {self.nazwa} ===")
        wynik = dane
        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
        print(f"Wynik końcowy: {wynik}")
        return wynik
    
# Tworzenie pipeline'a z komponentów
pipeline = Pipeline("Przetwarzanie danych sprzedaży")
pipeline.dodaj_krok(Czyszczenie())
pipeline.dodaj_krok(Normalizacja())
pipeline.dodaj_krok(Agregacja())

# Uruchomienie pipeline'a
dane_wejsciowe = [10, -5, 20, 30, -2, 40, 50]
wynik = pipeline.uruchom(dane_wejsciowe)
```

```python
# Przykład 2: Model z wbudowanymi komponentami

# Komponenty
class Preprocessor:
    def __init__(self):
        self.nazwa = "Preprocessor"

    def przetworz(self, dane):
        """Prostsze przetwarzanie – usunięcie None"""
        print(f"  [{self.nazwa}] Przetwarzam dane...")
        return [x for x in dane if x is not None]
    
class Ewaluator:
    def __init__(self):
        self.nazwa = "Ewaluator"

    def ocen(self, prawdziwe, przewidywane):
        """Oblicza błąd średniokwadratowy (MSE)"""
        print(f"  [{self.nazwa}] Obliczam MSE...")
        if len(prawdziwe) != len(przewidywane):
            return None
        mse = sum((p - t)**2 for p, t in zip(prawdziwe, przewidywane)) / len(prawdziwe)
        return mse
    
# Klasa Model używa kompozycji – zawiera Preprocessor i Ewaluator
class ModelML:
    def __init__(self, nazwa):
        """
        Model ML składa się z preprocessora i ewaluatora.
        """
        self.nazwa = nazwa
        self.preprocessor = Preprocessor()  # Kompozycja: ma preprocessor
        self.ewaluator = Ewaluator()        # Kompozycja: ma ewaluator
        self.parametry = None

    def trenuj(self, dane):
        """Symulacja treningu"""
        dane_czysty = self.preprocessor.przetworz(dane)
        print(f"  [Model {self.nazwa}] Trenuję na {len(dane_czysty)} obserwacjach...")
        # Prosta symulacja: parametr = średnia
        self.parametry = sum(dane_czysty) / len(dane_czysty)
        print(f"  Parametr modelu: {self.parametry:.2f}")

    def przewiduj(self, x):
        """Prosta predykcja"""
        if self.parametry is None:
            print("Model nie jest wytrenowany!")
            return None
        return self.parametry * x
    
    def ocen_model(self, prawdziwe, przewidywane):
        """Wykorzystuje wbudowany ewaluator"""
        return self.ewaluator.ocen(prawdziwe, przewidywane)
    
# Użycie
model = ModelML("Regresja Prosta")
model.trenuj([10, 20, None, 30, 40, None, 50])

predykcje = [model.przewiduj(1), model.przewiduj(2), model.przewiduj(3)]
prawdziwe = [30, 60, 90]
mse = model.ocen_model(prawdziwe, predykcje)
print(f"\nMSE: {mse:.2f}")
```

```python
# Przykład 3: Kompozycja vs Dziedziczenie – wybór podejścia

# PODEJŚCIE 1: Dziedziczenie (może prowadzić do problemów)
class ModelZPreprocessingiem:
    def preprocessuj(self, dane):
        return [x for x in dane if x > 0]
    
    def trenuj(self, dane):
        dane_czysty = self.preprocessuj(dane)
        print(f"Trenuję na {len(dane_czysty)} obserwacjach")

# PODEJŚCIE 2: Kompozycja (elastyczne, łatwiejsze w utrzymaniu)
class Preprocessor2:
    def preprocessuj(self, dane):
        return [x for x in dane if x > 0]
    
class ModelZKomponentami:
    def __init__(self, preprocessor):
        """
        Model przyjmuje preprocessor jako parametr – możemy podać dowolny!
        """
        self.preprocessor = preprocessor

    def trenuj(self, dane):
        dane_czysty = self.preprocessor.preprocessuj(dane)
        print(f"Trenuję na {len(dane_czysty)} obserwacjach")
        
# Użycie
preproc = Preprocessor2()
model = ModelZKomponentami(preproc)
model.trenuj([10, -5, 20, -3, 30])
```

# 🌱 Aspekt środowiskowy

```python
# Przykład: Efektywne przetwarzanie danych

# Komponent do próbkowania danych (zmniejsza zużycie zasobów)
class Probkowanie:
    def __init__(self, procent=10):
        """
        Komponent do próbkowania danych.
        Zmniejsza rozmiar datasetu, oszczędzając pamięć i czas obliczeń.
        """
        self.procent = procent
        self.nazwa = f"Próbkowanie {procent}%"

    def wykonaj(self, dane):
        """Zwraca tylko procent% danych"""
        n = max(1, len(dane) * self.procent // 100)
        print(f"  [{self.nazwa}] Redukuję dane z {len(dane)} do {n} obserwacji")

        # Prosta próbka co k-ty element
        krok = max(1, len(dane) // n)
        return dane[::krok]
    
# Pipeline z próbkowaniem dla oszczędności zasobów
pipeline_eco = Pipeline("Eco-Pipeline")
pipeline_eco.dodaj_krok(Probkowanie(procent=20))  # Tylko 20% danych
pipeline_eco.dodaj_krok(Normalizacja())

dane_duze = list(range(1, 101))  # 100 obserwacji
wynik_eco = pipeline_eco.uruchom(dane_duze)

# Porównanie: pipeline bez próbkowania przetwarzałby 100 obs,
# z próbkowaniem tylko 20 obs = 80% oszczędności zasobów
```

# 📊 Praca z Prawdziwymi Danymi

```python
# Dataset: Symulacja datasetu klientów

# Klasa reprezentująca pojedynczego klienta
class Klient:
    def __init__(self, id, imie, wiek, wydatki):
        """
        Pojedynczy klient w systemie.
        Parameters:
        - id: unikalny identyfikator
        - imie: imię klienta
        - wiek: wiek klienta
        - wydatki: lista wydatków (transakcji)
        """
        self.id = id
        self.imie = imie
        self.wiek = wiek
        self.wydatki = wydatki  # Lista kwot

    def suma_wydatkow(self):
        """Zwraca całkowitą sumę wydatków"""
        return sum(self.wydatki)
    
    def srednie_wydatki(self):
        """Zwraca średnie wydatki na transakcję"""
        if len(self.wydatki) == 0:
            return 0
        return self.suma_wydatkow() / len(self.wydatki)
    
    def __str__(self):
        """Reprezentacja tekstowa obiektu (wywoływana przez print())"""
        return f"Klient {self.id}: {self.imie}, {self.wiek} lat, wydał {self.suma_wydatkow()} zł"
    
# Klasa zarządzająca bazą klientów (kompozycja – ma listę obiektów Klient)
class BazaKlientow:
    def __init__(self):
        """Baza klientów jako lista obiektów Klient"""
        self.klienci = []

    def dodaj_klienta(self, klient):
        """Dodaje klienta do bazy"""
        self.klienci.append(klient)

    def znajdz_klienta(self, id):
        """Znajduje klienta po ID"""
        for klient in self.klienci:
            if klient.id == id:
                return klient
        return None
    
    def sredni_wiek(self):
        """Oblicza średni wiek wszystkich klientów"""
        if len(self.klienci) == 0:
            return 0
        return sum(k.wiek for k in self.klienci) / len(self.klienci)
    
    def top_klienci(self, n=5):
        """Zwraca n klientów z największymi wydatkami"""
        posortowani = sorted(self.klienci, key=lambda k: k.suma_wydatkow(), reverse=True)
        return posortowani[:n]
    
    def raport(self):
        """Generuje raport o bazie klientów"""
        print(f"=== Raport Bazy Klientów ===")
        print(f"Liczba klientów: {len(self.klienci)}")
        print(f"Średni wiek: {self.sredni_wiek():.1f} lat")
        print(f"Łączne wydatki: {sum(k.suma_wydatkow() for k in self.klienci)} zł")
        print(f"\nTop 3 klienci:")
        for i, klient in enumerate(self.top_klienci(3), 1):
            print(f"  {i}. {klient}")

# Tworzenie bazy klientów
baza = BazaKlientow()

# Dodawanie klientów
baza.dodaj_klienta(Klient(1, "Anna", 28, [120, 350, 80, 200]))
baza.dodaj_klienta(Klient(2, "Jan", 35, [500, 600, 450]))
baza.dodaj_klienta(Klient(3, "Katarzyna", 42, [100, 150, 200, 180, 220]))
baza.dodaj_klienta(Klient(4, "Piotr", 29, [80, 90, 70]))
baza.dodaj_klienta(Klient(5, "Maria", 51, [300, 400, 500, 600, 700]))

# Generowanie raportu
baza.raport()

# Wyszukiwanie klienta
klient = baza.znajdz_klienta(3)
if klient:
    print(f"\nZnaleziono: {klient}")
    print(f"Średnie wydatki na transakcję: {klient.srednie_wydatki():.2f} zł")

```

```python
# Analiza i Preprocessing z OOP

# Klasa do analizy danych klientów
class AnalizatorKlientow:
    def __init__(self, baza):
        """
        Analizator operuje na obiekcie BazaKlientow.
        """
        self.baza = baza

    def segmentuj_po_wieku(self):
        """Segmentacja klientów na grupy wiekowe"""
        mlodzi = []      # < 30
        sredni = []      # 30-50
        starsi = []      # > 50

        for klient in self.baza.klienci:
            if klient.wiek < 30:
                mlodzi.append(klient)
            elif klient.wiek <= 50:
                sredni.append(klient)
            else:
                starsi.append(klient)

        print("=== Segmentacja po wieku ===")
        print(f"Młodzi (<30): {len(mlodzi)} klientów")
        print(f"Średni (30-50): {len(sredni)} klientów")
        print(f"Starsi (>50): {len(starsi)} klientów")
        
        return {"mlodzi": mlodzi, "sredni": sredni, "starsi": starsi}
    
    def wykryj_vip(self, prog=1000):
        """Wykrywa klientów VIP (wydatki > prog)"""
        vip = [k for k in self.baza.klienci if k.suma_wydatkow() > prog]
        print(f"\n=== Klienci VIP (wydatki > {prog} zł) ===")
        for klient in vip:
            print(f"  {klient.imie}: {klient.suma_wydatkow()} zł")
        return vip
    
# Użycie analizatora
analizator = AnalizatorKlientow(baza)
segmenty = analizator.segmentuj_po_wieku()
vip = analizator.wykryj_vip(prog=1000)
```

```python
# Wizualizacja Eksploracyjna

# Pseudo-wizualizacja (bez matplotlib, używamy printu)
class WizualizatorKlientow:
    def __init__(self, baza):
        self.baza = baza

    def histogram_wydatkow(self):
        """Prosty histogram sumarycznych wydatków (tekstowy)"""
        print("\n=== Histogram Wydatków ===")
        
        for klient in sorted(self.baza.klienci, key=lambda k: k.suma_wydatkow()):
            bar = '*' * (klient.suma_wydatkow() // 100)  # 1 gwiazdka = 100 zł
            print(f"{klient.imie:12s} [{klient.suma_wydatkow():4d} zł] {bar}")

# Użycie wizualizatora
wizualizator = WizualizatorKlientow(baza)
wizualizator.histogram_wydatkow()
```
