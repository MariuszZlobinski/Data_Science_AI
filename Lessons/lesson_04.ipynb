```python
# Przykład 3: Operatory porównania i logiczne

# Operatory porównania zwracają True lub False
accuracy_model_1 = 0.92
accuracy_model_2 = 0.88
prog_akceptacji = 0.90

# Porównania
print(f"Model 1 lepszy niż Model 2: {accuracy_model_1 > accuracy_model_2}") # True
print(f"Model 1 osiągnął próg: {accuracy_model_1 >= prog_akceptacji}") # True
print(f"Model 2 poniżej progu: {accuracy_model_2 < prog_akceptacji}") # True
print(f"Modele równe: {accuracy_model_1 == accuracy_model_2}") # False
print(f"Modele różne: {accuracy_model_1 != accuracy_model_2}") # True

# Operatory logiczne: and, or, not
model1_ok = accuracy_model_1 >= prog_akceptacji
model2_ok = accuracy_model_2 >= prog_akceptacji

# and - oba warunki muszą być True
oba_modele_ok = model1_ok and model2_ok
print(f"Oba modele OK: {oba_modele_ok}") # False

# or - przynajmniej jeden warunek True
przynajmniej_jeden_ok = model1_ok or model2_ok
print(f"Przynajmniej jeden model OK: {przynajmniej_jeden_ok}") # True

# not - negacja
model2_wymaga_poprawy = not model2_ok
print(f"Model 2 wymaga poprawy: {model2_wymaga_poprawy}") # True

# Łączenie warunków - przykład z filtrowaniem danych
wiek_pacjenta = 45
cisnienie = 140

# Pacjent wysokiego ryzyka: wiek > 40 AND ciśnienie > 130
wysokie_ryzyko = (wiek_pacjenta > 40) and (cisnienie > 130)
print(f"Pacjent wysokiego ryzyka: {wysokie_ryzyko}") # True
```

# 2. Liczby całkowite i zmiennoprzecinkowe (int/float)

```python
# Przykład 1: Różnice między int i float

# Liczby całkowite
liczba_cech = 4 # int: liczba kolumn w datasecie Iris
liczba_klas = 3 # int: liczba kategorii (setosa, versicolor, virginica)
epoch = 100 # int: liczba epok treningowych

# Liczby zmiennoprzecinkowe
learning_rate = 0.001 # float: współczynnik uczenia
accuracy = 0.95 # float: dokładność modelu
sepal_length = 5.1 # float: długość działki kielicha (cm)

# Python automatycznie rozpoznaje typy
print(type(liczba_cech)) # <class 'int'>
print(type(learning_rate)) # <class 'float'>

# Konwersja między typami
wiek_float = 25.7
wiek_int = int(wiek_float) # konwersja na int - obcina część ułamkową
print(f"Wiek jako int: {wiek_int}") # 25 (nie zaokrągla, tylko obcina!)

liczba_probek = 150
liczba_jako_float = float(liczba_probek)
print(f"Liczba próbek jako float: {liczba_jako_float}") # 150.0

# Zaokrąglanie - użyj funkcji round()
temperatura = 23.7891
zaokraglona = round(temperatura, 2) # zaokrąglij do 2 miejsc po przecinku
print(f"Temperatura zaokrąglona: {zaokraglona}") # 23.79
```

```python
# Przykład 2: Operacje matematyczne w kontekście Data Science

import math # moduł z dodatkowymi funkcjami matematycznymi

# Obliczanie podstawowych statystyk (później użyjemy NumPy/Pandas)
# Przykład: ręczne obliczenie średniej dla 5 pomiarów temperatury
temp_1 = 22.3
temp_2 = 23.1
temp_3 = 21.8
temp_4 = 24.2
temp_5 = 22.6

# Średnia arytmetyczna
suma_temp = temp_1 + temp_2 + temp_3 + temp_4 + temp_5
liczba_pomiarow = 5
srednia = suma_temp / liczba_pomiarow
print(f"Średnia temperatura: {srednia:.2f}°C") # 22.80°C

# Obliczanie wariancji (uproszczone, bez NumPy)
# Wariancja = średnia kwadratów odchyleń od średniej
odch_1 = (temp_1 - srednia) ** 2
odch_2 = (temp_2 - srednia) ** 2
odch_3 = (temp_3 - srednia) ** 2
odch_4 = (temp_4 - srednia) ** 2
odch_5 = (temp_5 - srednia) ** 2

wariancja = (odch_1 + odch_2 + odch_3 + odch_4 + odch_5) / liczba_pomiarow
odchylenie_std = math.sqrt(wariancja) # pierwiastek z wariancji
print(f"Odchylenie standardowe: {odchylenie_std:.2f}°C") # ~0.84°C

# Przydatne funkcje matematyczne
wartosc = -15.7
print(f"Wartość bezwzględna: {abs(wartosc)}") # 15.7
print(f"Minimum: {min(temp_1, temp_2, temp_3, temp_4, temp_5)}") # 21.8
print(f"Maximum: {max(temp_1, temp_2, temp_3, temp_4, temp_5)}") # 24.2

# Funkcje z modułu math
print(f"Zaokrąglenie w górę: {math.ceil(srednia)}") # 23
print(f"Zaokrąglenie w dół: {math.floor(srednia)}") # 22
print(f"Logarytm naturalny: {math.log(liczba_pomiarow):.2f}") # 1.61
print(f"e^x: {math.exp(1):.2f}") # 2.72
```

```python
# Przykład 3: Precyzja float i problemy zaokrągleń

# Problem precyzji zmiennoprzecinkowej
a = 0.1
b = 0.2
suma = a + b
print(f"0.1 + 0.2 = {suma}") # 0.30000000000000004 - niespodzianka!
print(f"Czy suma == 0.3? {suma == 0.3}") # False!

# Rozwiązanie: porównywanie z tolerancją
epsilon = 1e-9 # bardzo mała wartość tolerancji
roznica = abs(suma - 0.3)
czy_rowne = roznica < epsilon
print(f"Czy suma ≈ 0.3? {czy_rowne}") # True

# Lepsze rozwiązanie: moduł math.isclose()
print(f"math.isclose(suma, 0.3): {math.isclose(suma, 0.3)}") # True

# Praktyczne zastosowanie w Data Science
# Porównywanie metryk modeli z tolerancją
accuracy_1 = 0.9234567
accuracy_2 = 0.9234568
roznica_accuracy = abs(accuracy_1 - accuracy_2)

if roznica_accuracy < 0.0001: # różnica mniejsza niż 0.01%
    print("Modele mają praktycznie identyczną dokładność")
else:
    print("Modele różnią się znacząco")

# Formatowanie wyników dla czytelności
loss_value = 0.003456789012
print(f"Loss (standardowo): {loss_value}")
print(f"Loss (2 miejsca): {loss_value:.2f}")
print(f"Loss (notacja naukowa): {loss_value:.2e}") # 3.46e-03
print(f"Loss (4 znaczące cyfry): {loss_value:.4g}") # 0.003457
```

```python
# Przykład 3: Formatowanie stringów - f-strings i metody format

# Dane przykładowe
model_type = "Logistic Regression"
accuracy = 0.9234
precision = 0.8891
recall = 0.9102
n_samples = 1500

# Sposób 1: Konkatenacja (przestarzały, nieczytelny)
raport_old = "Model: " + model_type + ", Accuracy: " + str(accuracy)
print(raport_old)

# Sposób 2: Metoda .format() (dobry, ale rozwlekły)
raport_format = "Model: {}, Accuracy: {:.2f}".format(model_type, accuracy)
print(raport_format)

# Sposób 3: f-strings (NAJLEPSZY, od Python 3.6+)
raport_fstring = f"Model: {model_type}, Accuracy: {accuracy:.2f}"
print(raport_fstring)

# F-strings z zaawansowanym formatowaniem
print(f"""
{'='*60}
RAPORT MODELU
{'='*60}
Model: {model_type}
Liczba próbek: {n_samples:,}
Accuracy: {accuracy:.2%}
Precision: {precision:.4f}
Recall: {recall:.4f}
F1-Score: {2 * precision * recall / (precision + recall):.4f}
{'='*60}
""")

# Wyjaśnienie formatowania:
# {accuracy:.2f} - float z 2 miejscami po przecinku
# {accuracy:.2%} - procenty z 2 miejscami (0.9234 -> 92.34%)
# {n_samples:,} - liczba z separatorem tysięcy (1500 -> 1,500)
# {precision:.4f} - float z 4 miejscami po przecinku
# {2*precision*recall/(precision+recall):.4f} - obliczenia wewnątrz fstringa!

# Wyrównywanie tekstu
headers = ["Model", "Accuracy", "Precision", "Recall"]
print(f"{headers[0]:<20} {headers[1]:>10} {headers[2]:>10} {headers[3]:>10}")
print(f"{model_type:<20} {accuracy:>10.2%} {precision:>10.2%} {recall:>10.2%}")
# < wyrównanie do lewej, > do prawej, ^ do środka
# liczba po : to szerokość pola

# Formatowanie dla ścieżek do plików
dataset_type = "train"
fold_number = 3
sciezka_output = f"data/processed/{dataset_type}_fold_{fold_number:02d}.csv"
print(f"Ścieżka: {sciezka_output}") # data/processed/train_fold_03.csv
# :02d oznacza liczbę całkowitą (d) z wypełnieniem zerami (0) do 2 cyfr
```
